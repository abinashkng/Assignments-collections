# Cleaning Dataset

```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
class Cleaner():
  def __init__ (self):
    pass
  def put_line_breaks(self,text):
    text = text.replace('</p>','</p>\n')
    return text
  def remove_html_tags(self,text):
      cleantext = bs(text, "lxml").text
      return cleantext

  def clean(self,text):
    text = self.put_line_breaks(text)
    text = self.remove_html_tags(text)
    return text
```

```python
cleaning = Cleaner()
df['cleaned_text'] = df[text_column_name].apply(cleaning.clean)
```
